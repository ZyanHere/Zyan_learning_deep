```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
```

```python
and_data['input1'] = [1,1,0,0]
and_data['input2'] = [1,0,1,0]
and_data['output'] = [1,0,0,0]
```

```python
xor_data['input1'] = [1,1,0,0]
xor_data['input2'] = [1,0,1,0]
xor_data['output'] = [0,1,1,0]
```

```python
and_data
```

```python
sns.scatterplot(x='input1', y='input2', hue='output', data=and_data, s=200)
```

```python
sns.scatterplot(x='input1', y='input2', hue='output', data=or_data, s=200)
```

```python
sns.scatterplot(x='input1', y='input2', hue='output', data=xor_data, s=200)
```

```python
from sklearn.linear_model import Perceptron

# Initialize perceptrons

clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
# Train perceptrons
clf1.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
clf3.fit(or_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
```

```python
clf1.coef_
```

```python
clf1.intercept_
```

```python
# Plot decision boundary for AND
x=np.linspace(-1,1,5)
y=-x+1
```

```python
plt.plot(x,y)
sns.scatterplot(x='input1', y='input2', hue='output', data=and_data, s=200)
```

```python
clf2.coef_
```

```python
clf2.intercept_
```

```python
# Create figure
plt.figure(figsize=(6, 6))

# Plot decision boundary for OR
x1 = np.linspace(-1, 1, 5)
y1 = -x1 + 0.5
plt.plot(x1, y1, label="OR Decision Boundary", color='red', linewidth=2)

# Plot OR scatter data
sns.scatterplot(x='input1', y='input2', hue='output', data=or_data, s=200, palette="viridis")

# Labels and legend
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.legend()
plt.title("OR Perceptron Decision Boundary")

# Show single plot
plt.show()
```

```python
# Create figure
plt.figure(figsize=(6, 6))

# XOR is not linearly separable, but we can visualize two possible regions
x1 = np.linspace(-0.5, 1.5, 5)
y1 = x1  # This is just a hypothetical boundary

# Plot two decision boundaries for XOR (simulating an MLP split)
plt.plot(x1, y1, label="Hypothetical XOR Boundary", color='red', linestyle="--", linewidth=2)
plt.plot(x1, -x1 + 1, label="Second Hypothetical Boundary", color='blue', linestyle="--", linewidth=2)

# Scatter plot for XOR dataset
sns.scatterplot(x='input1', y='input2', hue='output', data=xor_data, s=200, palette="viridis")

# Labels and legend
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.legend()
plt.title("XOR Dataset with Hypothetical Decision Boundaries")

# Show plot
plt.show()
```
